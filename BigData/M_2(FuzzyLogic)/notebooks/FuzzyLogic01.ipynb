```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt

MIN = 0
MAX = 75

# Create universe of discourse in Python using linspace ()
X = np.linspace(start=MIN, stop=MAX, num=75, endpoint=True, retstep=False)
X
```

```python
# FUSSIFICATION
MIDDLE = 50
interval1 = [MIN, MIDDLE/2, MIDDLE]
interval2 = [MIDDLE/2, MIDDLE, MAX]
interval3 = [MIDDLE, MAX, MAX]
young = fuzz.membership.trimf(X, interval1)
middle_aged = fuzz.membership.trimf(X, interval2)
old = fuzz.membership.trimf(X, interval3)
```

```python
plt.figure()
plt.plot(X, young)
plt.title("Young")
plt.grid(True)
```

```python
## Show the membership function for middle_aged
plt.figure()
plt.plot(X, middle_aged)
plt.title("Middle Aged")
plt.grid(True)
```

```python
plt.figure()
plt.plot(X, old)
plt.title("Old")
plt.grid(True)
```

```python
## Compute and show the COMPLEMENT of young
complement_young = 1 - young
plt.figure()
plt.plot(X, complement_young)
plt.title("Complement Young")
plt.grid(True)
```

```python
## Compute and show the UNION between young and middle_aged
young_or_middle = np.maximum(young,middle_aged)
plt.figure()
plt.plot(X, young_or_middle)
plt.title("Young or Middle")
plt.grid(True)
```

```python
## Compute and show the INTERSECTION between young and middle_aged
young_and_middle = np.minimum(young,middle_aged)
plt.figure()
plt.plot(X, young_and_middle)
plt.title("Young and Middle")
plt.grid(True)
```
